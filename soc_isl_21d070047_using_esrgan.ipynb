```python
import os
import matplotlib.pyplot as plt
import matplotlib.image as mpimg

# Function to display images from a directory
def display_images_from_directory(directory, title):
    images = [img for img in os.listdir(directory) if img.endswith(('.png', '.jpg', '.jpeg'))]
    num_images = len(images)

    plt.figure(figsize=(20, 20))

    for i, image_name in enumerate(images):
        img_path = os.path.join(directory, image_name)
        img = mpimg.imread(img_path)
        plt.subplot(1, num_images, i + 1)
        plt.imshow(img)
        plt.axis('off')
        plt.title(title)

    plt.show()

# Display images before super-resolution
display_images_from_directory('/content/ESRGAN/LR', 'Before Super-Resolution')

# Display images after super-resolution
display_images_from_directory('/content/ESRGAN/results', 'After Super-Resolution')
```
